```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Boston.csv')
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
plt.scatter(x=df['rm'],y=df['medv'],color='brown')
plt.xlabel("Avg. no. of rooms per dwelling")
plt.ylabel("Median value of home")
```

```python
#Getting the features and the target

X = pd.DataFrame(df['rm'])
y = pd.DataFrame(df['medv'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
DT.fit(X_train,y_train)
```

```python
y_pred = DT.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(y_test,y_pred))
print(mse)
```

```python
x = pd.DataFrame(df[['rm','lstat','age']])
Y = pd.DataFrame(df['medv'])
```

```python
x_train,x_test,Y_train,Y_test = train_test_split(x,Y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,Y_train)
```

```python
Y_pred = dt.predict(x_test)
```

```python
MSE = np.sqrt(mean_squared_error(Y_test,Y_pred))
print(MSE)
```

#### Second model is better than first model

## Decision Tree with IRIS dataset

```python
df1 = pd.read_csv('iris.csv')
```

```python
df1.head()
```

```python
X1 = pd.DataFrame(df1[['sepal_length','sepal_width','petal_length','petal_width']])
y1 = df1['species']
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT1 = DecisionTreeClassifier()
```

```python
DT1.fit(X1_train,y1_train)
```

```python
y1_pred = DT1.predict(X1_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y1_test,y1_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y1_test,y1_pred))
```
